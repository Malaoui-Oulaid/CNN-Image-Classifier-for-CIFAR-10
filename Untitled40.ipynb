```python
import tensorflow as tf
from keras import datasets
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, BatchNormalization, Dropout
from keras import models, layers
from matplotlib import pyplot as plt
```

```python
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
```

```python
print(train_images.shape)
print(test_images.shape)
```

```python
print(train_images[0].shape)
```

```python
plt.imshow(train_images[0])
```

```python
print(train_images[0][0][0])
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
print(train_images[0][0][0])
```

```python
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
print(train_labels[0])
```

```python
class_names[train_labels[0][0]]
```

```python
plt.imshow(train_images[100])
print(train_labels[100])
print(class_names[train_labels[100][0]])
```

```python
model1=Sequential()
model1.add(Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)))
model1.add(MaxPooling2D((2,2)))
model1.add(Conv2D(64,(3,3),activation='relu'))
model1.add(MaxPooling2D((2,2)))
model1.add(Conv2D(64,(3,3),activation='relu'))
model1.add(Flatten())
model1.add(Dense(64,activation='relu'))
model1.add(Dense(10,activation='softmax'))
```

```python
model1.summary()
```

```python
model1.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
h1=model1.fit(train_images,train_labels,epochs=30,shuffle=True,validation_data=(test_images,test_labels))
```

```python
result=model1.evaluate(test_images,test_labels)[1]
print(result)
```

```python
model2 = models.Sequential([
    layers.Conv2D(64, (3,3), activation='relu', padding='same', input_shape=(32,32,3)),
    layers.BatchNormalization(),
    layers.Conv2D(64, (3,3), activation='relu', padding='same'),
    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.25),

    layers.Conv2D(128, (3,3), activation='relu', padding='same'),
    layers.BatchNormalization(),
    layers.Conv2D(128, (3,3), activation='relu', padding='same'),
    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.35),

    layers.Conv2D(256, (3,3), activation='relu', padding='same'),
    layers.BatchNormalization(),
    layers.Conv2D(256, (3,3), activation='relu', padding='same'),
    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.45),

    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])
```

```python
model2.summary()
```

```python
model2.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
h2 = model2.fit(
    train_images, train_labels,
    epochs=5,
    batch_size=64,
    shuffle=True,
    validation_data=(test_images, test_labels)
)
```

```python
result2=model2.evaluate(test_images,test_labels)[1]
```
